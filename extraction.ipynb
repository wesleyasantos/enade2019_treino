## Extração dos Dados do Enade 2019

```python
# Importação das bibliotecas necessárias
import requests
import os
import pandas as pd
import numpy as np
import zipfile
from io import BytesIO
```

```python
# Criação de um diretório para armazenar o conteúdo do ENADE
os.makedirs('./enade2019', exist_ok = True)
```

```python
# Define a URL
url = 'https://download.inep.gov.br/microdados/microdados_enade_2019.zip'
```

```python
# Faz o download do conteúdo
filebytes = BytesIO(
    requests.get(url, verify = False).content
)

# Extrai o conteúdo do zipfile e salva na pasta enade2019
meuzip = zipfile.ZipFile(filebytes)
meuzip.extractall('./enade2019')
```

## Verificando os dados baixados

```python
enade = pd.read_csv('./enade2019/3.DADOS/microdados_enade_2019.txt', 
                    sep=';', 
                    decimal=',', 
                    encoding='utf-8',
                    low_memory=False,
)
```
